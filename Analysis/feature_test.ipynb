```python
import json
import pandas as pd
import re
file = "C:\\Users\\VORTEX\\Desktop\\prediction\\activiewvr-prediction-8c42c800f3b6\\cog1_wechs.json"
sol = []

with open(file, 'r') as f:
    data = json.load(f)
print(data)
df = pd.DataFrame.from_dict(data)
for col in list(df.columns):
    if re.search('',col):
        sol.append(col)
new_df = df[sol].dropna(how = 'all')  # dropping full nan row
new_cols = [col[:] for col in new_df.columns]
new_df.columns = new_cols
print(new_df)
new_df.to_csv(path_or_buf ='sjt.csv')
```
